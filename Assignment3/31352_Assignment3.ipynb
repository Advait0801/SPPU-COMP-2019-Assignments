```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
birth_record = pd.read_csv("birthwt.csv")

def covariance(dataset1,dataset2):
    mean1 = np.mean(dataset1)
    mean2 = np.mean(dataset2)
    print(np.sum(np.multiply(dataset1-mean1,dataset2-mean2))/len(dataset1))
    return np.sum(np.multiply(dataset1-mean1,dataset2-mean2))/len(dataset1)

def correlation(dataset1,dataset2):
    cov = covariance(dataset1,dataset2)
    sd1 = np.std(dataset1)
    sd2 = np.std(dataset2)
    print(cov/(sd1*sd2))
    return cov/(sd1*sd2)
```

```python
age = birth_record["age"]
age = age.to_numpy()

birthwt = birth_record["bwt"]
birthwt = birthwt.to_numpy()
correlation(age, birthwt)

plt.scatter(age,birthwt/1000,c ="blue")
plt.xlabel("Age")
plt.ylabel("Birth weight(Kg)")
```

# The correlation value is 0.9 and is very low, signalling the non-existent correlation in between maternal age and birth weight.

```python
lr = LinearRegression()
age = age.reshape(-1,1)
lr.fit(age,birthwt)

y = lr.predict(age)
print("Coefficients :",lr.coef_[0])
print("intercept :",lr.intercept_)

plt.plot(age,y,color= "red")
plt.scatter(age,birthwt,c= "blue")
plt.xlabel("Age")
plt.ylabel("Birth weight(g)")
plt.show()
```

# The correlation value is 0.18573328444909923 which is positive correlation, but the value is small which means the correlation is positive and small between the maternal weight and birth weight.

```python
motherswt = birth_record["lwt"]
motherswt = motherswt.to_numpy()

birthwt = birthwt/454
correlation(motherswt,birthwt)

plt.xlabel("Mother's weight")
plt.ylabel("Birth weight (Kg)")
plt.scatter(motherswt,birthwt,c = "blue")
```
