```python
ds['College'].fillna(ds['College'].mode()[0], inplace = True)
ds['Salary'].fillna(ds['Salary'].mean(), inplace = True)
ds
```

```python
ds.dropna(inplace = True)
ds
```

```python
age_groups = ds.groupby(ds['Age'])
age_groups
```

```python
age_groups['Salary'].describe()
```

```python
bins= [19,25,31,36,40]
labels = ['19-24','25-30','31-35','36-40']
ds['AgeGroup'] = pd.cut(ds['Age'], bins=bins,labels=labels, right=False)
```

```python
ds['AgeGroup'].value_counts()
```

```python
ds.groupby('AgeGroup')['Salary'].mean()
```

```python
ds.groupby('AgeGroup')['Salary'].median()
```

```python
ds.groupby('AgeGroup')['Salary'].describe()
```
