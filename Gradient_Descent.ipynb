```python
def gradient_descent(df, start_x, learning_rate, precision, max_iterations):
    """
    Performs gradient descent to find a local minimum of a function.

    Args:
        df: The derivative of the function to minimize.
        start_x: The initial value of x.
        learning_rate: The step size for each iteration.
        precision: The desired precision of the result.
        max_iterations: The maximum number of iterations.

    Returns:
        The value of x at the local minimum.
    """

    step_size = 1
    iterations = 0
    x = start_x
    gd = []

    print(f"Iteration: {iterations}, Value: {x}")
    while step_size > precision and iterations < max_iterations:
        prev_x = x
        x = x - learning_rate * df(prev_x)
        step_size = abs(prev_x - x)
        iterations += 1
        print(f"Iteration: {iterations}, Value: {x}")
        gd.append(x)

    return gd
```

```python
# Example usage
def my_derivative(x):
    return 2 * (x + 3)  # Derivative of (x + 3)^2


gd = gradient_descent(
    df=my_derivative,
    start_x=2,
    learning_rate=0.01,
    precision=0.00001,
    max_iterations=10000,
)

print("Local Minima Occurs at: ", gd[-1])
```
