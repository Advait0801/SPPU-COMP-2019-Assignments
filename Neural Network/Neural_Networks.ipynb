```python
import numpy as np
import pandas as pd
```

```python
ds = pd.read_csv('https://raw.githubusercontent.com/Advait0801/SPPU-COMP-2019-Assignments/refs/heads/ML/Neural%20Network/Churn_Modelling.csv')
ds
```

```python
ds = ds.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
ds
```

```python
gender = pd.get_dummies(ds['Gender'], drop_first=True)
geography = pd.get_dummies(ds['Geography'], drop_first=True)
ds = pd.concat([ds, gender, geography], axis=1)
ds = ds.drop(['Geography', 'Gender'], axis=1)
ds = ds.dropna()
ds
```

```python
X = ds.drop(['Exited'], axis=1)
Y = ds['Exited']
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(activation='relu', units=6))
model.add(Dense(activation='relu', units=6))
model.add(Dense(activation='sigmoid', units=1))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, batch_size=10, epochs=50)
```

```python
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)
y_pred
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_pred))
```
